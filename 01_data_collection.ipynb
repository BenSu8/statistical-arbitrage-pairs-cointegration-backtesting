```python
import yfinance as yf
import pandas as pd
import os
```

Pull the data for each of the tickers then create a dataframe with every ticker and only its adjusted close prices for the 8 years in the date range.

```python
pairs = [("MSFT", "AAPL"), ("KO", "PEP"), ("GS", "MS"), ("GM", "F")]

date_range = ["2018-05-28", "2026-05-29"]

all_prices = pd.DataFrame()

for pair in pairs:
    for ticker in pair:
        ticker_data = yf.download(ticker, start = date_range[0], end = date_range[1], auto_adjust = True)
        all_prices[f"{ticker} prices"] = ticker_data["Close"]

all_prices.head()
```

Save the prices dataframe into the current folder within data/raw.

```python
gen_dir = os.getcwd()
output_dir = os.path.join(gen_dir, "data", "raw")
file_name = "prices.csv"

os.makedirs(output_dir, exist_ok = True)

full_path = os.path.join(output_dir, file_name)

all_prices.to_csv(full_path, index = True)
```
